```python
import pandas as pd
import os
```

```python
folder_path = 'C:\\Users\\scarl\\Documents\\Research\\data\\'
stock_file = '80837.csv'

start_date = '1994-10-03'
end_date = '1994-10-07'
```

```python
full_path = os.path.join(folder_path, stock_file)
df = pd.read_csv(full_path)
df.head(5)
```

```python
column_to_filter = 'date'
df[column_to_filter] = pd.to_datetime(df[column_to_filter], format='%Y%m%d')
df.head(5)
```

```python
filtered_df = df[df[column_to_filter].between(start_date, end_date)]
filtered_df = filtered_df[['date', 'RETX']]
filtered_df.head(5)
```
